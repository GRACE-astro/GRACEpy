```python
def fill_submit_template(template_file,output_file,replacements):

    with open(template_file, "r") as f:
        content = f.read()

    for key, val in replacements.items():
        placeholder = f"@{key}@"
        content = content.replace(placeholder, str(val))

    with open(output_file, "w") as f:
        f.write(content)
```
